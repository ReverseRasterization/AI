```python
import requests
import random as rand
from bs4 import BeautifulSoup
```

```python
html = requests.get('https://www.scrapethissite.com/pages/simple/')
countries = []

if html.status_code == 200:
    soup = BeautifulSoup(html.text, 'html.parser')
    
    countries = [country.get_text(strip=True) for country in soup.find_all('h3', class_='country-name')]  
    
    print(countries)
    print(len(countries))
    
    a_count = 0
    for country in countries:
        if country[0] == 'A':
            a_count+=1
            
    print(a_count)
else:
    print(f"Failed! {html.status_code}")
```

```python
countries_clone = countries.copy()

def selectCountries():
    global countries_clone

    one, two, three = "", "", ""
    
    while (one == two) or (one == three) or (two == three):
        one, two, three = rand.choice(countries_clone), rand.choice(countries_clone), rand.choice(countries_clone)
    
    return one, two, three


while len(countries_clone) >= 3:
    
    c1, c2, c3 = selectCountries()
    
    if c1[0] == 'A' and c2[0] == 'A' and c3[0] == 'A':
        print(f"{3}/{len(countries_clone)}")
        
    countries_clone.remove(c1)
    countries_clone.remove(c2)
    countries_clone.remove(c3)
    
    

```

```python
import requests
import random as rand
import pandas as pd
from bs4 import BeautifulSoup

winsDF = pd.DataFrame()

def getWins(page_num):
    df = {
        'Team': [],
        'Wins': []
    }
    
    html = requests.get(f'https://www.scrapethissite.com/pages/forms/?page_num={page_num}')

    soup = BeautifulSoup(html.text, 'html.parser')

    teams = soup.find_all('tr', class_ = 'team')


    for team in teams:
        name = team.find('td', class_='name').get_text(strip=True)
        wins = int(team.find('td', class_='wins').get_text(strip=True))
        
        df['Team'].append(name)
        df['Wins'].append(wins)
    
    return pd.DataFrame(df)
        

for i in range(24):
    winsDF = pd.concat([winsDF, getWins(i)]) 

winsDF
```
