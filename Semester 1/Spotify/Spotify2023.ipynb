# Spotify Track Popularity Distribution

The minimum popularity score is 0 and the maximum is 100.
The median is a score of 45
25% of the scores lie within 0-24
50% of the scores lie within 25-45
75% of the scores lie within 46-62
The shape of our data is a symmetric, normal, bell curve

```python
import pandas as pd
import seaborn as sb

df = pd.read_csv('spotify2023.csv', encoding='latin-1')
```

```python
def calculate_mean(column):
    return column.sum()/len(column)

def calculate_stdv(column, stdevs):
    stdev = column.std()
    mean = column.mean()
    
    r1 = mean-(stdev * stdevs)
    r2 = mean+(stdev * stdevs)
    
    if r1 < column.min():
        r1 = column.min()
        
    if r2 > column.max():
        r2 = column.max()
        
    return [r1, r2]

df['streams'] = pd.to_numeric(df['streams'], errors = 'coerce')
calculate_stdv(df['streams'], 2)
```

```python
sb.histplot(data=df, x='danceability', bins=30)
```
